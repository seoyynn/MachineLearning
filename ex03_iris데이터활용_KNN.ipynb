### 데이터 전처리

```python
iris_data.data
```

```python
iris_df = pd.DataFrame(iris_data.data, columns = iris_data.feature_names)
iris_df.head(10)
```

```python
iris_df.head?
# ? -> 모에요? 물어보는 것
```

```python
# 문제와 답으로 분리
X = iris_df
y = iris_data.target
```

```python
# 훈련 데이터셋과 평과 데이터셋으로 분리
# train_test_split
from sklearn.model_selection import train_test_split
```

```python
# 문제, 답, 비율, 랜덤고정키값
# test_size = 0.3 => 30%만큼 평가 데이터로 분리
# random_state => 랜덤하게 섞이는 규칙을 고정
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 65)
X_train
```

```python
# 분리된 데이터 클래스(정답)별 개수 확인하기
# 7.5 : 2.5
import numpy as np
np.bincount(y_train)
# 데이터가 다양성과 형평성을 유지해야함 -> 학습에 도움됨
```

### 탐색적 데이터 분석(EDA)

```python
# 품종을 잘 분류해주는 기준은 어떤 컬럼인지
# 산점도 행렬 그래프
pd.plotting.scatter_matrix(X_train, figsize=(10,10), c = y_train, alpha = 0.9)
plt.show()
```

### 모델 선택 및 학습

```python
# 객체 생성
iris_knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
# fit(훈련문제, 훈련정답)
iris_knn_model.fit(X_train,y_train)
```

### 모델 평가

```python
# predict(테스트 문제)
pre = iris_knn_model.predict(X_test)
```

```python
# 정확도 계산 - accuracy_score
accuracy_score(pre, y_test)
```

### 하이퍼 파라미터 튜닝
- k값 범위 : 1~70
- train, test 정확도 값

```python
train_acc = []
test_acc = []

for i in range(1,71):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    train_acc.append(knn.score(X_train, y_train))
    test_acc.append(knn.score(X_test, y_test))
```

```python
plt.plot(range(1,71), train_acc, label='Train')
plt.plot(range(1,71), test_acc, label='Test')
plt.legend()
plt.show()
```

```python
test_acc_lst = []
train_acc_lst = []

for k in range(1,71):
    # 모델 생성 + 하이퍼 파라미터 지정
    knn_clf = KNeighborsClassifier(n_neighbors=k)
    knn_clf.fit(X_train,y_train)
   
    # test 정확도 저장
    test_pre = knn_clf.predict(X_test)
    test_score = accuracy_score(test_pre,y_test)
    test_acc_lst.append(test_score)
    
    # train 정확도 저장
    train_pre = knn_clf.predict(X_train)
    train_score = accuracy_score(train_pre,y_train)
    train_acc_lst.append(train_score)
```

```python
# 선 그래프

plt.figure(figsize=(10,4))

#테스트 정확도
plt.plot(range(1,71), test_acc_lst, label='test acc') 

#트레인 정확도
plt.plot(range(1,71), train_acc_lst, label='train acc') 

plt.grid() # 격자 표시
plt.xticks(range(1,71,2)) # X축 간격 조절
plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('acc score')

plt.show()
```
